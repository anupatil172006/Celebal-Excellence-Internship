# Handling missing Values

```python
df.isnull().sum()
```

```python
df["discount"]=df["discount"].fillna(df["discount"].mean())
```

```python
df["what_customers_said"]=df["what_customers_said"].fillna("No review")
df["seller_name"]=df["seller_name"].fillna("No seller name")
df["videos"]=df["videos"].fillna("No videos")
df["seller_information"]=df["seller_information"].fillna("No information")
df["variations"]=df["variations"].fillna("No variation")
```

```python
df.isnull().sum()
```

# Filtering Rows (pandas)

```python
high_rating=df[df["rating"]>4]
high_rating.head()
```

```python
high_discount=df[df["discount"]>50]
high_discount.head()
```

# Removing Duplicates

```python
df.duplicated().sum()
```

# Derived Column

```python
df[["initial_price","final_price"]].dtypes
```

```python
df["final_price"] = (
    df["final_price"]
    .str.replace('"', '', regex=False)
    .str.replace('‚Çπ', '', regex=False)
    .str.replace(',', '', regex=False)
)
```

```python
df["final_price"] = df["final_price"].astype(float)
```

```python
df["amount_saved"] = df["initial_price"] - df["final_price"]
```

```python
df[["initial_price","final_price","amount_saved"]].tail()
```

# Filtering Rows (pyspark)

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg
spark=SparkSession.builder \
    .appName("CEI_Task1") \
    .getOrCreate()
spark_df=spark.createDataFrame(df)
```

```python
spark_df.show(5)
```

```python
spark_df.printSchema()
```

```python
spark_df.groupBy("seller_name").count().show()
```

```python
from pyspark.sql.functions import avg

spark_df.groupBy("seller_name") \
        .agg(avg("rating").alias("average_rating")) \
        .show(20, truncate=False)
```
